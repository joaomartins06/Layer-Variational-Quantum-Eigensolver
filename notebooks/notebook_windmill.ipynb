```python
N_CLIQUES = 4
CLIQUE_SIZE = 5  # N_CLIQUES*(CLIQUE_SIZE-1)+1 = 17 nodes

G = nx.windmill_graph(N_CLIQUES, CLIQUE_SIZE)
print(f"n_nodes: {G.number_of_nodes()}, n_edges: {G.number_of_edges()}")
print(f"connected: {nx.is_connected(G)}")
```

```python
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(7, 7))
nx.draw(G, pos, with_labels=True, node_color="lightblue",
        node_size=400, font_size=10, font_weight="bold")
plt.title(f"Windmill graph: {N_CLIQUES} cliques of size {CLIQUE_SIZE} — "
          f"{G.number_of_nodes()} nodes, {G.number_of_edges()} edges")
plt.show()
```
